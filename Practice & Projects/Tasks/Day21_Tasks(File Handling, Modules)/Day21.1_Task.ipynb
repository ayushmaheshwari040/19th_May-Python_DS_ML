```python
header="Name,Math,Chemistry,Physics"
with open('student.csv','w') as f: 
    f.write(f"{header}\n")

rows=""
n=int(input("Enter no. of students: "))
for i in range(n): 
    print()
    name=input(f"Enter name of student{i+1}: ")
    maths=int(input("Enter marks in maths: "))
    chemistry=int(input("Enter marks in chemistry: "))
    physics=int(input("Enter marks in physics: "))
    rows+=f"{name},{maths},{chemistry},{physics}\n"
with open('student.csv','a') as f: 
    f.write(rows)
```

```python
with open('student.csv','r') as f: 
    data=f.readlines()

print(data)
```

```python
data[1:]
```

```python
data1=[]
for i in data[1:]: 
    a=i.strip().split(',')
    data1.append(a)

data1
```

```python
data2=[]
for i in range(len(data1)):
    data2_d={}
    s=sum(list(map(int,data1[i][1:])))
    data2_d['name']=data1[i][0]
    data2_d['totalmarks']=s
    data2.append(data2_d)

data2
```

```python
sorted_data2 = sorted(data2, key=lambda x: x['totalmarks'], reverse=True)
sorted_data2
```

```python
print("So, the top two Students are: ")
for i in range(2):
    print(sorted_data2[i])
```

```python
%%writefile bank_login_and_signup.py 

def login(acc_no, password): 
    with open("bank_data.csv","r") as f: 
        data=f.readlines()

    data=data[1:]

    data1=[]
    for i in range(len(data)): 
        a=data[i].strip().split(',')
        v=[a[0],int(a[1]),int(a[2]),int(a[3]),a[4]]
        data1.append(v)

    final_data_ls=[]
    for i in range(len(data1)): 
        final_data_dict={}
        final_data_dict['Name']=data1[i][0]
        final_data_dict['Balance']=data1[i][1]
        final_data_dict['Account Number']=data1[i][2]
        final_data_dict['Aadhar Number']=data1[i][3]
        final_data_dict['password']=data1[i][4]
        final_data_ls.append(final_data_dict)

    for user in final_data_ls:
        if user['Account Number'] == acc_no:
            if user['password'] == password:
                print(f"Welcome Back User! {user['Name']}")
                return True
            else:
                print("Invalid password.")
                return False
    else:            
        print("User does not exist.")


def signup(username, balance, password, aadhar): 
    import random
    acc_no=random.randint(1000,9999)

    row=f"{username},{balance},{acc_no},{aadhar},{password}\n"
    with open('bank_data.csv','a') as f: 
        f.write(row)

    print(f"Your data has be recorded and your account no. is: {acc_no}")
```
