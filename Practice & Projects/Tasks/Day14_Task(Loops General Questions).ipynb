# Q1.1

```python
num=int(input("Enter the number: "))

while num>=10:
    s=0
    while num>0:
        r=num%10
        s+=r
        num=num//10
        
    num=s    

print(f"The sum reduced to 1 digit is: {num} ")
```

# Q1.2

```python
a=int(input("Enter the number1 for HCF and LCM: "))
b=int(input("Enter the number2 for HCF and LCM: "))
print()
l=max(a,b)
s=min(a,b)

while l%s!=0: 
    r=l%s
    l=s
    s=r
    
print(f"The HCF or GCD of {a} and {b} is: {s}")
lcm=(a*b)//s                                           # Since lcm*hcf=n1*n2
print(f"The LCM of {a} and {b} is: {lcm}")
```

# Q3.

```python
total_valid_students = 0
high_total_count = 0
highest_total = -1
topper_roll = None

for i in range(1, 11):
    print(f"\nStudent {i}:")

    while True:
        roll = int(input("Enter roll number (1000-9999): "))
        if 1000 <= roll <= 9999:
            break
        else:
            print("Invalid roll number. Try again.")

    m1 = int(input("Enter marks in Subject 1: "))
    m2 = int(input("Enter marks in Subject 2: "))
    m3 = int(input("Enter marks in Subject 3: "))

    if m1 >= 40 and m2 >= 40 and m3 >= 40:
        total = m1 + m2 + m3
        total_valid_students += 1

        if total > 200:
            high_total_count += 1

        if total > highest_total:
            highest_total = total
            topper_roll = roll
    else:
        print("Student has marks < 40 in one or more subjects. Skipped.")

print("\n--- Summary ---")
print(f"Number of students with all marks  40: {total_valid_students}")
print(f"Number of students with total > 200: {high_total_count}")
if topper_roll:
    print(f"Roll number of student with highest total: {topper_roll} (Total: {highest_total})")
else:
    print("No valid student with all marks  40.")
```

# Q4

```python
num = int(input("Enter a number: "))

if num == 0:
    count = 1
else:
    count = 0
    num = abs(num)  # abs return absolute value for handling negative numbers
    while num > 0:
        num //= 10
        count += 1

print("Number of digits:", count)
```
