```python
def multipliers(): 
    return [lambda x : i * x for i in range(4)] 
print([m(2) for m in multipliers()]) 
```

# Q2.

```python
starts_with = lambda s, c: s.startswith(c)

string = input("Enter the string: ")
char = input("Enter the starting character to check: ")

if starts_with(string, char):
    print(f"Yes, the string starts with '{char}'.")
else:
    print(f"No, the string does not start with '{char}'.")
```

# Q3.

```python
print("Enter the digits in list1: ")
l1=list(map(int,input().split())) 
print("Enter the digits in list2: ")
l2=list(map(int,input().split())) 

print(f"""
The list l1 is: {l1}
The list l2 is: {l2}""")

intersection=lambda x, y: list(set(x).intersection(set(y))) 
print(f"The intersection of l1 and l2 is: {intersection(l1,l2)}")
```

# Q4.

```python
check=lambda x: True if sorted(x)==x else False

ls=[4,7,8,12,22,34]
if check(ls)==True: 
    print("The given list is sorted...")
else: 
    print("The given list is not sorted...")
```

# Q5.

```python
# Method1

add=lambda x,y,z: sum(x)+sum(y)+sum(z)

print("Enter the digits in list1")
ls1=list(map(int,input().split()))
print("Enter the digits in list2")
ls2=list(map(int,input().split()))
print("Enter the digits in list3")
ls3=list(map(int,input().split()))
print(f"""
List1 is: {ls1}
List2 is: {ls2}
List3 is: {ls3}""")

print(f"The sum of these three lists is: {add(ls1,ls2,ls3)}")
```

```python
# Method 2

List1=[1, 2, 3, 4]
List2=[10, 9, 8, 7]
List3=[1, 4, 7, 6]
sum_of_lists = map(lambda x, y,z: x+y+z, List1, List2, List3)
print(list(sum_of_lists)) # Output: [4, 10, 18]
```
