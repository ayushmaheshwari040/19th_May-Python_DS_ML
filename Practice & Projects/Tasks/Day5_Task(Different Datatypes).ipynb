### Shallow Copy
- Creates a new object, but does not copy nested objects; instead, it references them.
- Changes made to nested/mutable elements affect both copies.
- Syntax:-  new_obj = copy.copy(original_obj)            

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.copy(a)

b[0][0] = 100
print(a)  # [[100, 2], [3, 4]] → original is changed!
```

### Deep copy
- Creates a new object and recursively copies all nested objects.
- Changes in the copied object do not affect the original.
- Syntax:- new_obj = copy.deepcopy(original_obj)

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

b[0][0] = 100
print(a)  # [[1, 2], [3, 4]] → original is unchanged
```

# Q4.

1. Atomic data types or Primary data types:-
   - Basic built-in types that hold single, indivisible values. They are the fundamental building blocks of data.
   - Examples:- int, float, str, bool, complex
<br><br>
2. Secondary Data Types / User-Defined Data Types:-
   - Data types that are either collections of primary types or custom types created by the user.
   -  Examples:- list, tuple, dict, set, class Book()	

# Q5.

- A User-Defined Function (UDF) is a function that you create yourself in Python (or any programming language) to perform a specific task.
- Unlike built-in functions like print() or len(), UDFs are defined using the def keyword by the programmer.
  <br><br>
#
              Syntax:-  def function_name(parameters):
                            # block of code
                            return result

```python
def greet(name):
    return "Hello, " + name

print(greet("Ayush"))
```
