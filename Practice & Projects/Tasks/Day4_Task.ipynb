```python
# split()

text = "apple,banana,mango"
result = text.split(",")
print(result)
```

```python
# strip()

text1 = "   Hello World   "
text2 = "@@@Welcome@@@"

print(text1.strip())
print(text2.strip("@"))
```

```python
# replace()

text = "I like Java"
print(text.replace("Java", "Python"))
```

```python
# centre()

text = "Python"
print(text.center(20, "-"))
```

```python
# title()

text = "welcome to python programming"
print(text.title())
```

```python
# append()

fruits = ["apple", "banana"]
fruits.append("mango")
print(fruits)
```

```python
# pop()

numbers1 = [10, 20, 30, 40]
numbers2= [2,3,4,5,7]
numbers1.pop()
numbers2.pop(2)

print(numbers1)
print(numbers2)
```

```python
# remove()

colors = ["red", "blue", "green", "blue"]
colors.remove("blue")
print(colors)
```

```python
# sort()

nums1 = [3, 1, 4, 2]
nums2= [20,80,93,22,12]
nums1.sort()
nums2.sort(reverse=True)

print(nums1)
print(nums2)
```

    | **Mutable Data Types**                                | **Immutable Data Types**                         |
    | ----------------------------------------------------- | ------------------------------------------------ |
    | Can be changed or modified after creation             | Cannot be changed or modified once created       |
    | Object ID remains the same after modification         | A new object is created on modification          |
    | Examples: `list`, `dict`, `set`, `bytearray`          | Examples: `int`, `float`, `str`, `tuple`, `bool` |
    | Changes affect the original object                    | Changes do not affect the original object        |
    | Suitable for data that needs to be updated frequently | Suitable for data that should remain constant    |


# Q5.

- Identifiers:-
  - Identifiers are the names used to identify variables, functions, classes, modules, and objects in Python.
  - In simple words, whenever you give a name to something in Python, like a variable or function, you are creating an identifier.
  - Example:-    name = "Ayush"       # 'name' is an identifier
  <br><br>
- Rules for Writing Identifiers in Python:-
  - Can use letters (A-Z, a-z), digits (0-9), and underscores (_)
  - Cannot start with a digit
  - Cannot use special characters like @, $, %, #, etc.
  - Cannot be the same as Python keywords (e.g., class, def, if, etc.)
  - Case-sensitive — Name, name, and NAME are all different identifiers
  - No space allowed — use underscores (_) instead
