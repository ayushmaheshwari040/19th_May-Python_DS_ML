```python
def extendList(val, list=[]): 
    list.append(val) 
    return list 

list1 = extendList(10) 
list2 = extendList(123,[]) 
list3 = extendList('a') 
print("list1 = %s" % list1) 
print("list2 = %s" % list2) 
print("list3 = %s" % list3)
```

# Q2.

```python
def reverse(num):
    s=0
    while num>0:
        r=num%10
        s=s*10+r
        num=num//10
    return s

n=int(input("Enter the number: "))
print(f"The reverse of the number {n} is: {reverse(n)}")
```

# Q3.

```python
def palindrome(num):
    s=0
    while num>0:
        r=num%10
        s=s*10+r
        num=num//10
    return s

n=int(input("Enter the number: "))
n1=palindrome(n)

if n==n1:
    print(f"The no. {n} is a palindrome no.")
else:
    print(f"The no. {n} is not a palindrome no.")
```

# Q4.

```python
def isprime(num):
    if num==1:
        return 0
    else:
        for i in range(2, num):
            if num%i==0:
                return 0
        else:
            return 1

n=int(input("Enter no. for checking that it is prime or not."))
n1=isprime(n)

if n1==1:
    print(f"{n1}-> that means {n} is prime no.")
else:
    print(f"{n1}-> that means {n} is not prime no.") 
```

# Q5.

```python
def isprime(num):
    if num==1:
        return 0
    else:
        for i in range(2, num):
            if num%i==0:
                return 0
        else:
            return 1

s=int(input("Enter the starting no.: "))
e=int(input("Enter the ending no.: "))
print(f"The prime no.s from {s} to {e} are: ")

for i in range(s,e+1):
    if isprime(i)==1:
        print(i,end=" ")
```

# Q6.

```python
a = float(input("Enter the base (floating point number): "))
n = int(input("Enter the exponent (integer): "))
result = 1.0

if n > 0:
    for _ in range(n):
        result *= a
elif n < 0:
    for _ in range(-n):
        result *= a
    result = 1 / result
else:
    result = 1.0
    
print(f"{a} raised to the power {n} is: {result}")
```
