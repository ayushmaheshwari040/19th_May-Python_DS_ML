# Q1.

```python
text = input("Enter a string: ")

vowel_count = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0, 'A': 0, 'E': 0, 'I': 0, 'O': 0, 'U': 0}

for char in text:
    if char in vowel_count:
        vowel_count[char] += 1

print("\nVowel counts:")
for vowel, count in vowel_count.items():
    print(f"{vowel}: {count}")
```

# Q2.

```python
text = input("Enter a string: ")

characters = len(text)

total_words = len(text.split())

total_spaces = text.count(' ')

total_characters=characters-total_spaces

print(f"\nTotal Characters: {total_characters}")
print(f"Total Words: {total_words}")
print(f"Total Spaces: {total_spaces}")
```

# Q3.

```python
text = input("Enter a string: ")

punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
no_punct = ""

for char in text:
    if char not in punctuations:
        no_punct += char

print("\nString without punctuation:")
print(no_punct)
```

# Q5.

```python
# Function to input a matrix
def input_matrix(rows, cols, name):
    print(f"\nEnter elements for {name} matrix:")
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            val = int(input(f"Enter element [{i}][{j}]: "))
            row.append(val)
        matrix.append(row)
    return matrix

# Function to display a matrix
def display_matrix(matrix, name):
    print(f"\n{name} Matrix:")
    for row in matrix:
        print(row)

# Function to add two matrices
def add_matrices(A, B):
    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])
        result.append(row)
    return result

# Function to multiply two matrices
def multiply_matrices(A, B):
    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(B[0])):
            sum = 0
            for k in range(len(B)):
                sum += A[i][k] * B[k][j]
            row.append(sum)
        result.append(row)
    return result

# Function to transpose a matrix
def transpose_matrix(M):
    transposed = []
    for i in range(len(M[0])):
        row = []
        for j in range(len(M)):
            row.append(M[j][i])
        transposed.append(row)
    return transposed

# Input dimensions
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

# Input matrices
A = input_matrix(rows, cols, "First")
B = input_matrix(rows, cols, "Second")

# Display input matrices
display_matrix(A, "First")
display_matrix(B, "Second")

# Matrix Addition
if rows == len(B) and cols == len(B[0]):
    sum_matrix = add_matrices(A, B)
    display_matrix(sum_matrix, "Sum")
else:
    print("\nMatrix addition not possible (dimension mismatch).")

# Matrix Multiplication
if cols == len(B):
    product_matrix = multiply_matrices(A, B)
    display_matrix(product_matrix, "Product")
else:
    print("\nMatrix multiplication not possible (column of A != row of B).")

# Transpose
transpose_A = transpose_matrix(A)
transpose_B = transpose_matrix(B)
display_matrix(transpose_A, "Transpose of First")
display_matrix(transpose_B, "Transpose of Second")
```
