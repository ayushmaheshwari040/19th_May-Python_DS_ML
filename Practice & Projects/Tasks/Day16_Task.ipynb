# Q2.

```python
def prime_factors(n):
    factors=[]
    divisor=2
    while divisor<=n: 
        if n%divisor==0:
            factors.append(divisor)
            n=n/divisor
        else:
            divisor+=1
    return factors
    
num = int(input("Enter a number to find its prime factors: "))
factors = prime_factors(num)
print(f"Prime factors of {num} are: {', '.join(map(str, factors))}")
```

# Q3.

```python
def cubesum(n): 
    s=0
    while n>0: 
        r=n%10
        s=s+r**3
        n=n//10
    return s

num=int(input("Enter the no.: "))
print(f"The sum of cube of digits of {num} is: {cubesum(num)}")
```

# Q4

```python
def isleap(y):
    if (y%4==0 and y%100!=0) or y%400==0: 
        return 1
    else:
        return 0

year=int(input("Enter the year for checking the leap year: "))
leap=isleap(year)
if leap==1:
    print(f"{leap} i.e {year} is leap year.")
else:
    print(f"{leap} i.e {year} is not leap year.")
```

```python
def sumOfProperDivisior(n):
    sumlst=[]
    for i in range(1,n): 
        if n%i==0:
            sumlst.append(i)
            
    return sum(sumlst)

num1=int(input("Enter the num1: "))
num2=int(input("Enter the num2: "))
s1=sumOfProperDivisior(num1)
s2=sumOfProperDivisior(num2)
if s1==num2 and s2==num1: 
    print(f"{num1} and {num2} are amicable number...")
else: 
    print(f"{num1} and {num2} are not amicable number...")
            
```
