- Types of data structures:-
  1. Array
      - Image processing 
      - Storing sensor data 
      - Static lists (e.g., calendar dates)
  2. Linked list
      - Music playlists 
      - Undo/redo functionality 
      - Memory management in OS
  3. Stack 
      - Browser history 
      - Expression evaluation 
      - Backtracking algorithms (e.g., maze solving) 
  4. Queue

      - CPU scheduling 
      - Call center systems
  5. Tree 
      - Database indexing

      - Password verification
  6. Hash Table 
      - Database indexing 
      - Caching (e.g., DNS cache) 
      - Password verification
  7. BST 
      - Search operations 
      - Sorted map implementation
  8. Heap 
      - Task scheduling
      - Dijkstra's algorithm (shortest path) 
      - Real-time simulations

    | **Ordered Data Type**                   | **Unordered Data Type**             |
    | --------------------------------------- | ----------------------------------- |
    | Maintains insertion order               | Does not maintain order             |
    | Can be accessed using index             | Cannot be accessed using index      |
