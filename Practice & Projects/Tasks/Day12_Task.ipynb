# Q3.

 Consider a list (list = []). You can perform the following commands:
1. insert i e: Insert integer e at position i .
2. print: Print the list.
3. remove e: Delete the first occurrence of integer e .
4. append e: Insert integer e at the end of the list.
5. sort: Sort the list.
6. pop: Pop the last element from the list.
7. reverse: Reverse the list.
  <br>
Initialize your list and read in the value of n followed by n lines of commands where each command will be of the 7 types listed above. Iterate through each command in order and perform the corresponding operation on your list.

```python
lst=[1,2,3,4,5]
print(f"Initially the list is: {lst}")
operations="""" 
                1. Enter 1 for insert element at specific poisition.
                2. Enter 2 for print the list.
                3. Enter 3 for remove the element from the list at first occurence.
                4. Enter 4 for to append the element at last.
                5. Enter 5 for sort the list.
                6. Enter 6 for pop last item from the list.
                7. Enter 7 for reverse the list.
                8. Enter number out of 1 to 7 for exit the operations. """

while True:
    v=int(input(f"{operations} \n"))
    if v==1:
        p=int(input("Enter the poisition: "))
        e=int(input("Enter the number you want to insert: "))      
        lst.insert(p,e) 
        
    elif v==2:
        print(lst)
        
    elif v==3:
        e=int(input("Enter the number you want to remove: "))
        if e not in lst:
            print("The number is not present in the list...")
        else:
            lst.remove(e)

    elif v==4:
        e=int(input("Enter the number you want to append: "))
        lst.append(e)
        
    elif v==5:
        lst.sort()

    elif v==6:
        lst.pop()

    elif v==7:
        lst.reverse()
    else:
        break
```

# Q4. 

The provided code stub will read in a dictionary containing key/value pairs of name:[marks] for a list of students. Print the average of the marks array for the student name provided, showing 2 places after the decimal.

```python
student={}
n=int(input("Enter no. of students you want to insert: "))
print()
for i in range(n):
    name=input(f"Enter name of student{i+1}: ")
    marks=[]
    for j in range(3):
        m=int(input(f"Enter marks in subject{j+1} out of 20: "))
        marks.append(m)
    student[name]=marks
    print()

print(student)
print()

s=input("Enter the student name from the given dictionary for average marks of that student: ")
avg=sum(student[s])/3
print(f"The average marks of {s} is: {avg:.2f}")
```

# Q5.

### a. Bubble Sort

```python
lst=[34,17,20,10,33,1,4,6]

for i in range(len(lst)-1):
    for j in range(i+1,len(lst)):
        if lst[i]>lst[j]:
            temp=lst[i]
            lst[i]=lst[j]
            lst[j]=temp
            
print(lst)            
```

### b. Insertion Sort

```python
lst=[34,17,20,10,33,1,4,6]

for i in range(1, len(lst)):
    key = lst[i]
    j = i - 1
    while j >= 0 and key < lst[j]:
        lst[j + 1] = lst[j]
        j -= 1
    lst[j + 1] = key

print(lst)    
```

### c. Selection Sort

```python
lst=[34,17,20,10,33,1,4,6]

n = len(lst)
for i in range(n):
    min_index = i
    for j in range(i + 1, n):
        if lst[j] < lst[min_index]:
            min_index = j
    lst[i], lst[min_index] = lst[min_index], lst[i]

print(lst)
```
