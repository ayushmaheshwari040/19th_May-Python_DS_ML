```python
def batting(p): 
    batl=[]
    for i in range(len(p)): 
        batd={}
        batscore=p[i]['runs']//2
        if p[i]['runs']>=50: 
            batscore+=5
        elif p[i]['runs']>=100: 
            batscore+=10
        elif ('balls' in p[i]) and 80 <= ((p[i]['runs']/p[i]['balls'])*100) <= 100: 
            batscore+=2
        elif ('balls' in p[i]) and ((p[i]['runs']/p[i]['balls'])*100) > 100: 
            batscore+=4
        elif ('4' and '6') in p[i]:
            batscore+=p[i]['4']
            batscore+=2*(p[i]['6']) 
        batd['name']=p[i]['name']
        batd['batscore']=batscore
        batl.append(batd)
    return batl

def bowling(p): 
    bowll=[]
    for i in range(len(p)): 
        bowld={}
        bowlscore=0
        if 'wkts' in p[i]: 
          bowlscore=p[i]['wkts']*10
        elif ('wkts' in p[i]) and p[i]['wkts']>=3: 
            bowlscore+=5
        elif ('wkts' in p[i]) and p[i]['wkts']>=5: 
            bowlscore+=10
        elif ('overs' in p[i]) and 3.5<=(p[i]['runs']/p[i]['overs'])<=4.5: 
            bowlscore+=4
        elif ('overs' in p[i]) and 2<=(p[i]['runs']/p[i]['overs'])<=3.5:
            bowlscore+=7
        elif ('overs' in p[i]) and (p[i]['runs']/p[i]['overs'])<2:
            bowlscore+=10
        bowld['name']=p[i]['name']
        bowld['bowlscore']=bowlscore
        bowll.append(bowld)
    return bowll

def fielding(p): 
    fieldl=[]
    for i in range(len(p)):
        fieldd={}
        fieldscore=0
        if (p[i]['field']==1) and ('catch' or 'stumping' or 'runout') in p[i]: 
            fieldscore=10
        fieldd['name']=p[i]['name']
        fieldd['fieldscore']=fieldscore
        fieldl.append(fieldd)
    return fieldl        

player=[ {'name':'Virat Kohli', 'role':'bat', 'runs':112, '4':10, '6':0, 'balls':119, 'field':0}, 
         {'name':'du Plessis', 'role':'bat', 'runs':120, '4':11, '6':2, 'balls':112, 'field':0}, 
         {'name':'Bhuvneshwar Kumar', 'role':'bowl', 'wkts':1, 'overs':10, 'runs':71, 'field':1}, 
         {'name':'Yuzvendra Chahal', 'role':'bowl', 'wkts':2, 'overs':10, 'runs':45, 'field':0}, 
         {'name':'Kuldeep Yadav', 'role':'bowl', 'wkts':3, 'overs':10, 'runs':34, 'field':0} ]
```

```python
batting_data=batting(player)
batting_data
```

```python
bowling_data=bowling(player)
bowling_data
```

```python
fielding_data=fielding(player)
fielding_data
```

```python
final_data=[]
for i in range(len(player)):
    if player[i]['role']=='bat': 
        final_data.append(batting_data[i])
    elif player[i]['role']=='bowl': 
        final_data.append(bowling_data[i])

final_data
```
