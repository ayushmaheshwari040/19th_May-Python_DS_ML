```python
from getpass import getpass
import random

# password_validation function
def password_validation():

    while True:
        i=0
        lower=False
        upper=False
        digit=False
        special=False    
        password=getpass("Enter password: ")
        if len(password)>=8:
            while i<len(password):
                if password[i].islower():
                    lower=True
                elif password[i].isupper():
                    upper=True
                elif password[i].isdigit():
                    digit=True
                elif not password[i].isidentifier():
                    special=True
                i+=1
    
            if all([lower,upper,digit,special]):
                #print(f"Your Password '{password}' is valid...")
                return password
            else:
                print(f"Your Password '{password}' not fulfil condtions so, is not valid ...")
                continue
        else: 
            print("Length of password should be greater than equal to 8...  ")
            continue



# signup Function
def signup():
    username=input("Enter username: ")         #username            
    password=password_validation()          #valid_password
    print()
    
    while True:
        intial_amt=int(input("Enter the initial amount: "))          #initial_amt        
        if intial_amt>=20000:                                            
            acc_no=random.randint(1000,9999)          #acc_no    
            print(f"Your Rs.{intial_amt} has been added into your account and the account no. is: {acc_no}")
            break
        else:
            print("Warning: Your amount should be atleast 20000 !")

    user_details.append({
        'user_name':username,
        'user_password':password,
        'user_initial_amt':intial_amt,
        'user_acc_no':acc_no
    })
    print()
        

# login function
def login():
    acc_no=int(input("Enter account number: "))
    password=password_validation()
    
    for i in range(len(user_details)):
        if user_details[i]['user_acc_no']==acc_no and user_details[i]['user_password']==password:
            print("You are you are loggedin to your account...")
            print()
            while True:
                n=int(input("""
                            1. Press 1 for credit into account
                            2. Press 2 for debit from account
                            3. Press 3 for changing the password
                            4. Press 4 for check the balance
                            5. Press 5 for show details
                            6. Press 6 for Logout\n"""))
                if n==1:
                    amt=int(input("Enter the amount you want to credit: "))
                    user_details[i]['user_initial_amt']+=amt
                    
                elif n==2:
                    amt=int(input("Enter the amount you want to debit: "))
                    if amt<=user_details[i]['user_initial_amt']:
                        user_details[i]['user_initial_amt']-=amt  
                        
                elif n==3:
                    print("Current Password: ")
                    current_password=password_validation()
                    print()
                    if current_password == user_details[i]['user_password']:
                        print("New Password: ")
                        new_password=password_validation()
                        user_details[i]['user_password']=new_password
                        print("Notice: Your Password has been changed")    
                    else:
                        print("Warning: Your Current Password is Invalid !")

                elif n==4:
                    print(f"Your balance is: Rs.{user_details[i]['user_initial_amt']}")

                elif n==5:
                    print(f"Your username is: {user_details[i]['user_name']}  and account no. is: {user_details[i]['user_acc_no']}")

                elif n==6:
                    print()
                    return
    else:
        print("Warning: Your account doesn't exist .")       
        print()

# forgot password function
def forgotpassword():
    username=input("Enter username: ")
    acc_no=int(input("Enter account no.: "))

    for i in range(len(user_details)):
        if user_details[i]['user_name']==username and user_details[i]['user_acc_no']==acc_no:
            print("New Password: ")
            new_password=password_validation()
            user_details[i]['user_password']=new_password
            print()
            print("Your Password has been successfully changed.....")
            print()
            return
    else:
        print("Warning: Your username or account no. is incorrect or doesn't exist! Please try next time.")
        print()


#--------------------------------------------------------------------------------------------------------------------------------------------#

print("  WELCOME TO MY BANK  ".center(132,"*"))
print("\n\n")
user_details=[]          #for signup function
i=0

while True:
    print(f"  Phase{i+1}  ".center(100,"-"))
    n=int(input("""
               1. Press 1 For Login
               2. Press 2 For Signup
               3. Press 3 If you Forgot Password
               4. Press 4 For Exit.\n"""))
    if n==1:
        login()
    elif n==2:
        signup()
    elif n==3:
        forgotpassword()
    elif n==4:
        print(" THANK YOU ! We appreciate your visit. ")
        break 
    i+=1
    
```
