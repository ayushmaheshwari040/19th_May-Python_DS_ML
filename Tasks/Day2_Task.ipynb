# Q4.

- Compiler
  - Translates the entire program into machine code before execution.
  - Faster, as the whole code is translated at once.
  - Detects all errors after scanning the whole code.
  - Produces a separate executable file.
  - High memory usage  stores compiled code in memory.
  - Harder to debug, as all errors are shown after compilation.
  - Preferred for performance-critical applications.
  - Examples: C, C++, Java (with JIT), Go
  <br><br>
- Interpreter
  - Translates one line at a time into machine code and executes it.
  - Slower, as it processes code line by line.
  - Detects errors line-by-line; stops at the first error.
  - No separate output file; executes directly.
  - Low memory usage  doesnt store code after execution.
  - Easier debugging, as errors are shown step-by-step.
  - Preferred for scripting and rapid development.
  - Examples: Python, JavaScript, Ruby, PHP

- What is a Data Structure?
  - A data structure is a way of organizing and storing data in a computer so that it can be used efficiently.
  - It defines how data is arranged, accessed, manipulated, and stored for better performance and easier coding.
  - Examples of Common Data Structures:- stack, queue, dictionary, set, tree, graph
  <br><br>   
-  Why Do We Need Data Structures in Programs?
   - Efficient Access
   - Organized Data Management
   - Performance Optimization
   - Logical Representation
   - Reusability & Modularity
   - Foundation of Algorithms
