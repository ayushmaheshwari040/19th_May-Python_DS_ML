# 
    | **Dictionary**                                          | **List**                                               |
    | ------------------------------------------------------- | ------------------------------------------------------ |
    | Stores data in **key-value** pairs                      | Stores data as an **ordered sequence of elements**     |
    | Keys are **unique and unordered**                       | Elements are **indexed and ordered**                   |
    | Accessed using **keys** (`my_dict['key']`)              | Accessed using **index** (`my_list[0]`)                |
    | Useful for **mapping relationships** (e.g., name  age) | Useful for **ordered collections** (e.g., to-do items) |
    | Defined with **curly braces `{}`**                      | Defined with **square brackets `[]`**                  |

# Q8.

- Escape characters in Python are special characters used to represent certain whitespace or special characters inside strings that are otherwise hard to include directly.
They start with a backslash \ followed by a character that has a special meaning.
###
    | Escape Character | Meaning              | Example                                             |
    | ---------------- | -------------------- | --------------------------------------------------- |
    | `\n`             | Newline (line break) | `"Hello\nWorld"`  prints on two lines              |
    | `\t`             | Horizontal tab       | `"Hello\tWorld"`  inserts a tab space              |
    | `\\`             | Backslash character  | `"C:\\Users\\Name"`  prints `C:\Users\Name`        |
    | `\'`             | Single quote         | `'It\'s sunny'`  includes `'` inside string        |
    | `\"`             | Double quote         | `"He said, \"Hello\""`  includes `"` inside string |
    | `\r`             | Carriage return      | Moves cursor to the beginning of the line           |
    | `\b`             | Backspace            | Removes previous character (in some cases)          |
