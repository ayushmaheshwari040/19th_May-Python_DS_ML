```python
import requests

def place_serach(query, lat, lon, sort="RELEVANCE", limit=10):
    url = "https://places-api.foursquare.com/places/search?"
    
    data={
        'query':'hotels',
        'll':f"{lat},{lon}",
        'sort':sort,
        'limit':limit
    }
    
    headers = {
        "accept": "application/json",
        "X-Places-Api-Version": "2025-06-17",
        "authorization": "Bearer 3N30GDUB5MAHXBGAP5S3KQ5WZU1KGT11N2SVNU14GBHNIEJ3"
    }
    
    resp = requests.get(url, params=data, headers=headers)
    if resp.status_code==200: 
        d=resp.json()
        ls=d['results']
        for i in ls: 
            name=i['name']
            address=i['location']['formatted_address']
            print(f"Place name: {name}")
            print(f"Place address: {address}")
            print()
    else: 
        print(f"Error occured Status code: {resp.status_code} Reason: {resp.reason}")
```

```python
place_serach('hotels', 26.9124,75.7873, limit=50)
```
