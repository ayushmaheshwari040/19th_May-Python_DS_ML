###  random library-  
#####                **Task:- No. guessing game**

```python
# Task- No. guessing game

import random

computer_think_num=random.randint(1,50)
i=0
while i<5:
   user_guess=int(input(f"Chance {i+1} Enter your guess between 0 to 50 ")) 
   if user_guess<1 or user_guess>50:
       print("Warning: Please enter no. in 0 to 50")
       continue
    
   if user_guess==computer_think_num:
       print("The guessed no. is correct")
       break
   elif user_guess<computer_think_num:
       print("Hint: Think big no.")
   else:
       print("Hint: Think small no. ")          
   i+=1
    
else:
    print("You loose your all chance")
```
