# File Handling

- What is file ?
  <br>
  File is a container on a computer system that stores data, such as text, images, or program code

- Types of files:-
  1. Human Readable
  2. Machine Readable

- Extension of files:-
  .txt, .ppt, .mkv, .mp4, .csv, .jpg  etc.

- File System:-
  - Data stored in **memory blocks**.
  - Group of multiple blocks is **sector**.
  - Blocks are present in **partition table**.
  - Each block has **inode no**.
  - **FAT(file allcoation table )** file system is used before.
  - **NTFS(new technology file system)** currently used in windows.

- Types of Paths:-
  1. Absolute- from root
  2. Relative- from current

### 'w' mode
- override the existing file or clear data.
- new file creates if does not exist.

```python
file=open('new.txt','w')
file
```

```python
file.write('Hello hii bye ')
```

```python
file.write('Hello hii bye')
file.close()
```

```python
file=open('new.txt','w')
file
```

```python
file.write('New data added')
file.close()
```

```python
f=open('data.txt','w')
f.writelines(['1. rohan','2. aman','3. ram','4. rajat'])
f.close()
```

```python
f=open('data.txt','w')
f.writelines(['1. rohan\n','2. aman\n','3. ram\n','4. rajat'])
f.close()
```

### 'a' mode
- new file creates if not exist
- append the data in existing data without overriding

```python
f=open('data.txt','a')
f.writelines(['5. ayush\n', '6. sachin','7. gaurav'])
f.close()
```

```python
f=open(table.txt','w')

n=int(input("Enter no. to print table into file: "))
for i in range(10):
    f.write(f"{n} * {i+1} = {n*(i+1)}\n")

print(f"Table of {n} is in the file.")
f.close()
```

### with keyword
- auto close file instead of manually.
- context manager

```python
with open('new.txt', 'r') as file:          # alias is file work as object                    
    data=file.read()                        # operation in this block
    print(data)
```

```python
file.read()
```
